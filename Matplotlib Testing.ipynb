# Matplotlib Introduction

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint = True)
C, S = np.cos(X), np.sin(X)
```

```python
plt.plot(X,C)
plt.plot(X,S)
plt.show()
```

```python
#create a figure of size 8x6 inches, 80 dots per inch
plt.figure(figsize=(8,6), dpi = 80)

#create a new subplot from a grid of 1x1
plt.subplot(1,1,1)
#plot cosine with a blue continuous line of width 1 pixel
plt.plot(X,C, color = 'blue', linewidth=1.0, linestyle='-')
#plot sine with a green continuous line of width 1 pixel
plt.plot(X,S, color = 'green', linewidth=1.0, linestyle = '-')
#set x limits
plt.xlim(-4,4)
#set x ticks
plt.xticks(np.linspace(-4,4,9, endpoint = True))
#set y limits
plt.ylim(-1,1)
#set y ticks
plt.yticks(np.linspace(-1,1,5, endpoint=True))
#save figure using 72 dots per inch
#plt.savefig("exercise_2.pgn", dpi = 72)
plt.show()
```

```python
#create a figure of size 8x6 inches, 80 dots per inch
plt.figure(figsize=(8,6), dpi = 80)
#changing color and widths
plt.plot(X,C, color = 'blue', linewidth=2.5, linestyle='-', label='cosine')
plt.plot(X,S, color = 'red', linewidth=2.5, linestyle = '-', label='sine')
#setting limits
plt.xlim(X.min()*1.1, X.max()*1.1)
plt.ylim(C.min()*1.1, C.max()*1.1)
#setting ticks
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
#setting tick labels
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], 
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1,0,+1], [r'$-1$', r'$0$', r'$+1$'])

#moving spines
ax = plt.gca()  # gca stands for 'get current axis'
ax.spines['right'].set_color('none') #remove the box
ax.spines['top'].set_color('none') #remove the box
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) #top to bottom axis
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0)) #left to right axis

#adding a legend
plt.legend(loc='upper left') #label attribute

#annote some points
t = 2/3 * np.pi
#connect 2 points to form a line (t, 0), (t, cos(t)) using linewidth or linestyle
plt.plot([t,t], [0, np.cos(t)], color = 'blue', linewidth =2.5, linestyle='--')
plt.plot(t, np.cos(t), marker='*', markersize=20, color ='blue')

plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',
            xy=(t,np.cos(t)), xycoords ='data',
            xytext=(-90,-50), textcoords='offset points', fontsize =16,
            arrowprops =dict(arrowstyle ='->', connectionstyle='arc3, rad=.2'))

plt.plot([t,t], [0, np.sin(t)], color='red', linewidth=2.5, linestyle='--')
plt.scatter([t], [np.sin(t)], s=1000, marker='$This$', color = (1,0,1))

plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$', xy=(t, np.sin(t)),
            xycoords ='data', xytext=(+10,+30), textcoords = 'offset points', fontsize = 16,
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2'))

#Make the tick labels more visiable
for label in ax.get_xticklabels()+ ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='white', edgecolor = 'None', alpha =0.65))

plt.show()
```

# Basic Text Commands

```python
fig = plt.figure()
fig.suptitle('bold figure suptitle', fontsize=14, fontweight='bold')

ax = fig.add_subplot(1,1,1) # nRow, nColumn, nthFigure
fig.subplots_adjust(top=0.85)
ax.set_title('axes title')

ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')

#text (location (x,y), text)
ax.text(3, 8, 'boxed italics text in data coords', style='italic',
        bbox={'facecolor':'red', 'alpha':0.5, 'pad':10})

#r'$LaTeX$', style (bold,italic), fontsize
ax.text(2, 6, r'an equation: $E=mc^2$', fontsize=15)

#unicode
ax.text(3, 2, u'unicode: Institut f\374r Festk\366rperphysik')

#alignment, red/green/black/alpha rgba 0-1
ax.text(0.95, 0.01, 'colored text in axes coords',
        verticalalignment='bottom', horizontalalignment='right',
        transform=ax.transAxes,
        color='green', fontsize=15)


ax.plot([2], [1], 'o')
#annotate (text, xy location, xytext location, )
ax.annotate('annotate', xy=(2, 1), xytext=(3, 4),
            arrowprops=dict(facecolor=(1,0,0,1), shrink=0.05))

#axis = xlim (min, max), ylim(min, max)
ax.axis([0, 10, 0, 10])

plt.show()
```

# Many Examples

## Filled Plots

```python
import numpy as np
import matplotlib.pyplot as plt

n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint = True) #min, max, nPoints
Y = np.sin(2 *X)

plt.axes([0.025, 0.025, 0.95, 0.95])

#fill_between (domain X, y1(could be mid point), y2 (function), where function <>)
plt.plot(X, Y+1, color = 'blue', alpha = 1.00)
plt.fill_between(X, 1, Y+1, color = 'blue', alpha = .25)

plt.plot(X, Y-1, color = 'blue', alpha = 1.00)
plt.fill_between(X, -1, Y-1, where = (Y-1)> -1, color = 'blue', alpha = .25)
plt.fill_between(X, -1, Y-1,  where = (Y-1)<-1, color = 'red', alpha = .25)

plt.xlim(-np.pi, np.pi)
#plt.xticks(()) #no x ticks
plt.ylim(-2.5, 2.5)
#plt.yticks(()) #no y ticks

plt.show()
```

## Scatter Plot

```python
n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)

#make 1024 random colors
Colors = np.random.random(n) 

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.scatter(X, Y, s=75, c=Colors, alpha= 0.6  )

plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)

plt.show()
```
